```python
import pandas as pd

# Load CSV
df = pd.read_csv("aapl_2022.csv")

# Quick check
df.head()  # shows first 5 rows
```

```python
# First 5 rows
print(df.head())

# Last 5 rows
print(df.tail())

# Quick info about columns, types, and non-null counts
print(df.info())

# Summary statistics for numeric columns
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values (if any)
df = df.dropna()

# OR fill missing numeric values with mean
# df['Close'] = df['Close'].fillna(df['Close'].mean())
```

```python
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
for col in numeric_cols:
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)
```

```python
print(df.head(10))
```

```python
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')

# Drop rows where any numeric column is NaN (conversion failed)
df = df.dropna(subset=numeric_cols)
```

```python
print(df.head())
print(df.info())
```

```python
df.to_csv("aapl_2022_cleaned.csv", index=False)
```

```python
# Check basic statistics
print(df.describe())

# Check unique values in categorical columns (if any)
# print(df['ColumnName'].unique())

# Plot Closing Price trend
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("AAPL Closing Price Trend")
plt.show()
```
